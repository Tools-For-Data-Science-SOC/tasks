We draw the graph of the given data set with log(no of cases ) vs number of days 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)

# Preprocessing Input data
data = pd.read_csv('task1.csv')
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
y = np.log(Y)
plt.scatter(X, y)
plt.show()
```


First we have to use linear regression using gradient descent algorithm to determine the hypothesis function for the given data set.


```python
# Building the model
m = 0
c = 0

L = 0.0001  # The learning Rate
epochs = 5000  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    y_pred = m*X + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (y - y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(y - y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
    
print (m, c)
```

now we plot the hypothesis funcion (log of number of cases vs number of days) for the the given data so that we can predict the no of cases at any given day.

```python
# Making predictions
y_pred = m*X + c
plt.rcParams['figure.figsize'] = (12.0, 9.0)

plt.scatter(X, y) 
plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color='red')  # regression line
plt.show()
```


Now we plot the graph of number of cases vs number of days

```python
Y_pred = pow(np.e,(m*X + c))
plt.rcParams['figure.figsize'] = (12.0, 9.0)

plt.scatter(X, Y) 
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.show()
```
